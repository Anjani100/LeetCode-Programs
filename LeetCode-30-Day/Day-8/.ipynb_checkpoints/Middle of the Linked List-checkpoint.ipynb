```python
class Node:
    
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    
    def __init__(self):
        self.head = None
        self.tail = None
        
    def createList(self, llist):
        print('Enter -1 to stop adding numbers')
        while True:
            s = int(input('Enter the number you want to insert: '))
            if(s == -1):
                break
            if(llist.head == None):
                llist.head = Node(s)
                llist.tail = llist.head
                llist.head.next = None
                llist.tail.next = None
            else:
                temp = Node(s)
                llist.tail.next = temp
                llist.tail = temp
                llist.tail.next = None
        return llist
    
    def ins_beg(self, llist):
        s = int(input('Enter the data you want to insert at the beginning: '))
        temp = Node(s)
        temp.next = llist.head
        llist.head = temp
        return llist
    
    def ins_end(self, llist):
        if(llist.head == None):
            print('List is empty. First create one, Idiot!')
            llist = llist.createList(llist)
            return llist
        s = int(input('Enter the data you want to insert at the end: '))
        temp = Node(s)
        temp.next = None
        llist.tail.next = temp
        llist.tail = temp
        return llist
    
    def ins_after(self, llist):
        if(llist.head == None):
            print('List is empty. First create one, Idiot!')
            llist = llist.createList(llist)
            return llist
        val = int(input('Enter the data after which you want to insert: '))
        temp = llist.head
        flag = 0
        while(temp):
            if(temp.data == val):
                flag = 1
            temp = temp.next
        if(flag == 0):
            print('The data after which you wanna insert is not there!!!')
            return llist
        s = int(input('Enter the data you want to insert after the value: '))
        num = Node(s)
        temp = llist.head
        while(temp.data != val):
            temp = temp.next
        num.next = temp.next
        temp.next = num
        return llist
    
    def printList(self):
        temp = self.head
        while(temp):
            print(temp.data, end=" -> ")
            temp = temp.next
        print('X')

if __name__=='__main__':
    llist = LinkedList()
    while True:
        print('1. Create a Linked List.')
        print('2. Display the Linked List.')
        print('3. Insertion at the beginning.')
        print('4. Insertion at the end.')
        print('5. Insertion after a value.')
        print('Enter -1 to exit.')
        ch = int(input('Enter your choice: '))
        if(ch == 1):
            llist = llist.createList(llist)
        elif(ch == 2):
            llist.printList()
        elif(ch == 3):
            llist = llist.ins_beg(llist)
        elif(ch == 4):
            llist = llist.ins_end(llist)
        elif(ch == 5):
            llist = llist.ins_after(llist)
        elif(ch == -1):
            break
```
